```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("Sample.xlsx")
```

```python
a = df['Churn'].value_counts()
plt.figure(figsize=(4,2))
bars = plt.bar(a.index, a.values, color=['r','g'])
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.title("Customer Churn Distribution")
plt.bar_label(bars, padding=1)
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.hist(df['tenure'], bins=11, color='b')
plt.xlabel("Tenure")
plt.ylabel("Customers")
plt.title("Tenure Distribution")
plt.show()
```

```python
a = df.groupby('sim')['MonthlyCharges'].mean()
plt.figure(figsize=(4,2))
bars = plt.bar(a.index, a.values,color=['yellow','r','skyblue','b'])
plt.xlabel("SIM")
plt.ylabel("Average Monthly Charges")
plt.title("Average Monthly Charges per SIM")
plt.bar_label(bars,padding=-9)
plt.show()
```

```python
df['Tenure_Quarter'] = (df['tenure'] // 3) + 1
quarter_counts = df['Tenure_Quarter'].value_counts().sort_index()
plt.figure(figsize=(5,3))
plt.bar(quarter_counts.index, quarter_counts.values)
plt.xlabel("Tenure (Quarters)")
plt.ylabel("Number of Customers")
plt.title("Customer Tenure Distribution (Quarterly)")
plt.show()
```
